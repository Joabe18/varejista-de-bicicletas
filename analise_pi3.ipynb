### Subtotal

```python
#@title 12. Subtotal
amostra['subtotal'].value_counts()
```

```python
amostra['subtotal'].describe()
```

```python
# Calcula as estatísticas descritivas específicas
subtotal_media = amostra['subtotal'].mean()
subtotal_mediana = amostra['subtotal'].median()
subtotal_minimo = amostra['subtotal'].min()
subtotal_maximo = amostra['subtotal'].max()
subtotal_desvio_padrao = amostra['subtotal'].std()
subtotal_primeiro_quartil = amostra['subtotal'].quantile(0.25)
subtotal_terceiro_quartil = amostra['subtotal'].quantile(0.75)
subtotal_iqr = subtotal_terceiro_quartil - subtotal_primeiro_quartil
subtotal_variancia = amostra['subtotal'].var()
subtotal_skewness = amostra['subtotal'].skew()
subtotal_kurtosis = amostra['subtotal'].kurtosis()
subtotal_contagem = amostra['subtotal'].count()

# Exibe as estatísticas descritivas específicas com 2 casas decimais
print(f"Média de subtotal: {subtotal_media:.2f}")
print(f"Mediana de subtotal: {subtotal_mediana:.2f}")
print(f"Mínimo de subtotal: {subtotal_minimo:.2f}")
print(f"Máximo de subtotal: {subtotal_maximo:.2f}")
print(f"Desvio padrão de subtotal: {subtotal_desvio_padrao:.2f}")
print(f"Primeiro Quartil (25%): {subtotal_primeiro_quartil:.2f}")
print(f"Terceiro Quartil (75%): {subtotal_terceiro_quartil:.2f}")
print(f"Intervalo Interquartil (IQR): {subtotal_iqr:.2f}")
print(f"Variância de subtotal: {subtotal_variancia:.2f}")
print(f"Skewness de subtotal: {subtotal_skewness:.2f}")
print(f"Kurtosis de subtotal: {subtotal_kurtosis:.2f}")
print(f"Contagem de observações de subtotal: {subtotal_contagem:.2f}")
```
